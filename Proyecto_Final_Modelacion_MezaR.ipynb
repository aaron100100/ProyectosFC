## Introducción

La **dinámica de poblaciones** es una rama científica dedicada al estudio de las poblaciones biológicas, abordando aspectos como tamaño, dimensiones físicas, estructuración por edad y sexo, y otros parámetros definitorios. Utilizando ecuaciones matemáticas, esta disciplina modela el comportamiento de las poblaciones con el objetivo de prever cambios numéricos, identificar causas, anticipar comportamientos y analizar las consecuencias ecológicas.

Este campo, fundamental en biología matemática, ha encontrado aplicaciones significativas en la gestión de recursos biológicos, evaluación de impactos ambientales de actividades humanas y en investigaciones médicas relacionadas con infecciones y dinámicas celulares.

Dentro de la dinámica de poblaciones, existen modelos que varían en complejidad técnica y en su capacidad para reflejar la realidad observada. Estos modelos pueden ser discretos o continuos. Los modelos discretos consideran el estado del sistema solo en instantes de tiempo específicos, utilizando ecuaciones recursivas, mientras que los modelos continuos, basados en ecuaciones diferenciales, permiten considerar el estado del sistema en cualquier momento.

Aunque aparentemente simples, los modelos discretos han demostrado su utilidad en diversos campos, especialmente en la dinámica de poblaciones. La primera sección aborda modelos unidimensionales que se centran en la evolución de una única población. Las siguientes secciones tratan modelos multidimensionales que involucran interacciones entre varias poblaciones, utilizando sistemas de ecuaciones lineales o no lineales, dependiendo de la complejidad del fenómeno estudiado.


Este proyecto presenta el **Modelo de Beverton-Holt** que estudia poblaciones de tiempo discreto a partir de una ecuación en diferencias. Presentamos el modelo, $$x_{n+1}=\frac{ax_n}{1+bx_n},$$donde $a>0$ representa la tasa de reproducción y $b>0$ es la mortalidad. 
Basándose fuertemente en el contexto de pesca por **Beverton Y Holt**. Este trabajo se derivó bajo las suposiciones de  competición de concurso o también llamado competición de Torneo que este último tiene las siguientes caracteristicas; hay individuos más fuertes que otros, y los más fuertes siempre tienen recursos y se pueden reproducir sin problemas, mientras que los más débiles obtienen menos recursos y se reproducen menos, o incluso mueren. Por lo tanto, siempre hay individuos reproduciendose implicando que haya un número de crias al menos constante. De esta forma una ayuda visual del modelo es que la gráfica siempre es creciente y tiende asintoticamente a un valor máximo cuando  $x\rightarrow\infty$.

### Primer Gráfica del Modelo 

```python
#Parámetros
# a = representa la tasa de reprodución 
#b = La tasa de mortalidad
a = 10
b = 2

values_x = np.linspace(0,20,1000)
model_values = a*values_x/(1 + b*(values_x))
x_valores_fijos = [0,(a-1)/b]

plt.plot(values_x,model_values)
plt.xlabel("$x_n$")
plt.ylabel("$x_{n+1}$")
#plt.ylim([0,5])
plt.title('Modelo de Beverton-Holt')
plt.grid(True)
plt.show()
```

En nuestro Modelo nuestra función de reproducción es la siguiente:
$$f(x) =  \frac{ax}{1 + bx } $$

Vamos a verificar que es creciente y tiende asintoticamente a un valor máximo cuando $x $ se va a infinito, procedemos a obtener la derivada de $f$

Rapidamente,
$$ f'(x)  = \frac{a(1 + bx) - ax(b)}{(1 + bx)^{2}} = \frac{a + abx - abx}{(1 + bx)^{2}} = \frac{a}{(1 + bx)^{2}} $$
De  esta forma podemos afirmar que $f$ es creciente. Ya solo nos falta ver que pasa cuando hacemos el límite de $x$ cuando tiende a infinito.
$$\lim_{x\rightarrow\infty}f(x)= \lim_{x\rightarrow\infty}\frac{ax}{1 + bx }$$
Por la regla de l'Hopital concluimos que:

$$\lim_{x\rightarrow\infty}f(x)= \lim_{x\rightarrow\infty}\frac{ax}{1 + bx } = \frac{a}{b}$$
Con estas dos observaciones podemos garantizar que el Modelo se trata de una competición Torneo. A partir de esto podemos ir estableciendo una afirmación que tiene sentido Biologico, La asintota significa que habrá un número máximo de individuos para la siguiente generación, no importa que tan grande sea la población inicial, siempre va a parecer constante la población.


### Presentamos graficos de La Dinámica Poblacional con el Modelo De Beverton-Holt
#### Con distintas condiciones iniciales


```python
#Aqui creamos una función que nos va a ayudar a graficar el modelo 
def f(x,a,b):
    return (a*x)/(1 + b*x)
```

```python
#Condición incial, se empieza con 5 individuos
x0=[5]
#Declaramos los valores de los parametros a y b 
a = 20 # Tasa de reproducción
b = 0.02 # Tasa de Mortalidad

#Iteramos la función 51 veces
for i in range(50):
    #Añadimos el valor de g()
    x0.append(f(x0[-1],a,b))
#Graficamos estos valores (x1) como puntos ('.') ya que es una ecuación discreta
plt.plot(x0,'.')
#Etiquetamos ejes:
plt.xlabel('$n$')
plt.ylabel('$x_n$')
plt.title('Evolución de la Población. ci = 5, a = 20, b = 0.02')
plt.grid()
plt.show()
```

A continuación, vamos a calcular de forma analitica los puntos de equilibrio o puntos fijos de nuestro Modelo, estos son de importancia para la modelación por que rapidamente podemos conocer si se va a mantener constante la población de una determinada especie al cabo de unos años o si se van a extinguir con, siempre y cuando conozcamos la población inicial.

Vamos a obtenerlos de la siguiente manera:

$$f(x^{*}) = x^{*}$$
Y esto nos dice que igualemos nuestra función $f$ a la función Identidad.

$$ f(x^{*}) = x^{*} \implies \frac{ax^{*}}{1 + bx^{*} } = x^{*}$$
De aquí podemos obtener nuestro primer punto fijo que es $x^{*} = 0$. Para obtener el siguiente vamos a suponer que $x^{*}$ es distinto de cero.
$$ \frac{ax^{*}}{1 + bx^{*} } = x^{*} \implies  ax^{*} = x^{*}\left( 1 + bx^{*}\right) \implies ax^{*}  = x^{*} + b(x^{*})^{2}  \implies x^{*} + b(x^{*})^{2} - ax^{*} = 0 \implies b(x^{*})^{2} + (1 - a)x^{*} = 0\implies x^{*}\left(bx^{*} + (1-a)\right) = 0$$
Y como $x^{*}\neq 0$ entonces debe pasar que $bx^{*} + (1-a) = 0$ de esta última ecuación despejamos a $x^{*}$
$$bx^{*} = a -1 \implies x^{*} = \frac{a-1}{b} $$ El último despeje es válido por que $b > 0$. Por lo tanto los puntos fijos para nuestro Modelo de Dinámica Poblacional son:

- $x^{*} = 0$
- $x^{*} = \frac{a-1}{b}$


Ahora graficando el modelo, la Identidad y los puntos de equilibrio se ve de la siguiente forma:

```python
#Parámetros
# a = representa la tasa de reprodución 
#b = La tasa de mortalidad
a = 10
b = 2

values_x = np.linspace(0,20,1000)
model_values = a*values_x/(1 + b*(values_x))
x_valores_fijos = [0,(a-1)/b]

plt.plot(values_x,model_values)
plt.plot(values_x, values_x)
plt.scatter(x_valores_fijos, x_valores_fijos, color = 'hotpink' )
plt.xlabel("$x_n$")
plt.ylabel("$x_{n+1}$")
plt.ylim([-0.5,5])
plt.xlim([-0.5, 5])
plt.title('Modelo de Beverton-Holt')
plt.grid(True)
plt.show()
```

Lo siguiente en nuestro estudio del Modelo, es poder obtener la estabilidad de los puntos de equilibrio obtenidos. De esta forma podemos concluir cual es el cantidad maxima de la población o en que momento se llegará a la extinción.

Un punto de equilibrio puede ser estable o inestable, y se establece de acuerdo al siguiente criterio:

- Si $\left| f'(x^{*})\right|< 1$ entonces $x^{*}$ es estable.
- Si $\left| f'(x^{*})\right|> 1$ entonces $x^{*}$ es inestable.

Aplicando este criterio, llegamos a las siguientes estabilidades, recordando que,
$$f'(x) = \frac{a}{(1 + bx)^{2}} $$ con $a,b > 0$

Para $x^{*} = 0$, evaluamos:
$$ |f'(0)| = |a| $$ de esta manera tomando $a \in (0, 1)$ concluimos que $x^{*} = 0$ es estable y si $a \in(1,\infty)$ entonces $x^{*} = 0$ es inestble. Pasamos al siguiente punto de equilibrio.

Para $x^{*} = \frac{a-1}{b}$, evaluamos:
$$ \left|f'\left(\frac{a-1}{b}\right)\right| = \left|\frac{a}{\left(1 + b\left(\frac{a-1}{b}\right)\right)^{2}}\right|  = \left|\frac{a}{a^{2}} \right| = \frac{1}{|a|}$$

Haciendo la siguiente elección vamos a tener que:

- Si $a\in(1,\infty)$ entonces $\frac{1}{|a|} < 1$ por lo tanto $x^{*} = \frac{a-1}{b}$  es estable.
- Si $a\in(0,1)$ entonces $\frac{1}{|a|} > 1$ por lo tanto $x^{*} = \frac{a-1}{b}$  es inestable.

### Analizando un caso en particular

Vamos a estudiar una población de insectos que crece de acuerdo al modelo de Beverton- Holt. Con condición inicial de 200 

```python
#Condición incial, se empieza con 200 individuos
x0=[200]
#Declaramos los valores de los parametros a y b 
a = 20 # Tasa de reproducción
b = 0.02 # Tasa de Mortalidad

#Iteramos la función 51 veces
for i in range(50):
    #Añadimos el valor de g()
    x0.append(f(x0[-1],a,b))
#Graficamos estos valores (x1) como puntos ('.') ya que es una ecuación discreta
plt.plot(x0,'.', color = 'hotpink')
#Etiquetamos ejes:
plt.xlabel('$n$')
plt.ylabel('$x_n$')
plt.title('Evolución de la Población')
plt.grid(True)
plt.show()
```

LLegamos a las siguientes conclusiones, el modelo va a tener una capacidad maxima de individuos que se obtiene facilmente de la siguiente forma:
$$ M = \frac{a - 1}{b} = \frac{20-1}{0.02} = 950$$
Biologicamente entendemos esto como sigue: **La asintota significa que hay un número máximo en la siguiente generación, no importa que tan grande sea la población inicial**
Y por el analisis previo de estabilidad tenemos lo siguiente:

- $(0,0)$ es inestable
- $(950,950)$ es estable

Justificación:
$$f'(x) = \frac{20}{(1 + 0.02x)^{2}} $$

- $|f'(0)| = |20| = 20 > 1 $ y entonces 0 es punto de equilibrio inestable
- $|f'(950)| = |\frac{1}{20}| = \frac{1}{20} < 1$ entonces 950 es punto de equilibrio estable.

Podemos concluir que bajo los supuestos siguientes, tasa de reproducción constante $a = 20$ y tasa de mortalidad $b =  0.02 $ la población de insectos va a alcanzar un capacidad máxima de población, nunca se van a extinguir por esta cuestión de que el punto de equlibrio $(0,0)$ es inestable. Por otro lado $(950,950)$ es estable por que no importa que condición inicial eligamos las soluciones vas a converger monotonamente a $M$.
