Una vez en el ecosistema, los microorganismos en el suelo y los sedimentos del depósito destruyen el mercurio y producen un producto químico muy tóxico conocido como metilmercurio. 

El mercurio experimenta un proceso llamado bioacumulación, que ocurre cuando los organismos absorben contaminantes más rápido que lo que tardan sus cuerpos en eliminarlos. ##### Planteamiento del Problema

Para presentar el modelo de acumulación de metilmercurio en un adulto vamos a tomar lo siguiente a consideración. La ecuación en diferencias que nos va a describir como evoluciona la acumulación de esta sustancia en un adulto va a hacer los cálculos cada 70 días, por lo tanto para poder calcular el consumo de peces por semana y mejor aun la concentración de metilmercurio debemos hacer una conversion. Nuestro ecuación en diferencias está dada por la siguiente:
$$ \Delta a_{n}  = ka_{n} + C$$
De la informacipon que nos proporciona el enunciado la tasa de decremento del metilmercurio en un adulto es del 50%, de esta forma nuestro valor de $k$ es 0.5. Para calcular $C$ haremos lo siguiente. Dado que nuestro modelo hace cálculos cada 70 días debemos saber que tantos peces con esta sustancia consume un adulto en 70 días. Sabemos que un adulto consume 2 peces cada semana y cada pez en promedio pesa 0.817 kg y el valor medio de metilmercurio en cada pez es de 0.43 microgramos por gramo. Haciendo las operaciones necesarias llegamos a que: 
$$ C = 20(0.817 kg)\left(\frac{0.43 μ g}{g}\right) = 20(817g)\left(\frac{0.43μ g}{g}\right) = 7026.2 μ g = 7.0262 mg $$
Multiplicamos por 20 pues esos son el número de peces que se consumen en 70 días. Por lo tanto nuestro ecuación en diferencias es:

$$ a_{n + 1} = 0.5 a_{n} + C$$

Otra cosa importante es que supondremos  una condición inicial $a_{0} = 0$, estto quiere decir que en principio el adulto promedio tiene nada de concentración de mercurio en el cuerpo.

Ya con esto, comenzamos a escribir nuestro programa.

```python
'''
Algo que se debe notar es que no estamos poniendo tal cual el valor de C obtenido arriba debido a que en el inciso siguiente
debemos ir cambiando el valor de peces, es por eso que:
l = es el producto de 817g con 0.43 microgramos/g 
p = número de peces
'''
#Condición Inicial
a_0 = 0.0 
#Esta variable representa el producto de  817g con 0.43 microgramos/ g
l = 0.35131
#Número de peces
p = 20
#Tasa de decremento
K = 0.5
metilMercurio = []
'''
Supondremos que la vida promedio de un persona es de 70 años
70 años equivale a 840 meses, y esto a su vez a 3360 semanas, finalmente en dias son 23520 días, diviendo entre 70 días obtenemos que
23250 / 70 = 332.1428.... como esta es a la iteración a la que debemos llegar para completar 70 años vamos a tomar la parte entera, es decir 332
'''
#Comenzamos con un ciclo for que se repite 333 veces
for i in range(333):
    if (i == 0): # este if nos sirve para agregar nuestra condición iniial
        metilMercurio.append(a_0)
    else:  # en este else vamos agregando un elemento gracias a que ya conocemos la tasa de información del elemento anterior
        metilMercurio.append(K*metilMercurio[i - 1] + l*p)
```

 - Ahora, suponga que se ha estado pagando la hipoteca por $8$ años, y tenemos una oportunidad de refinanciar la deuda. Vamos a considerar que el pago mensual es de $\frac{250,000}{360} = 694.444$ esto sin que se cuente el interés. El problema nos dice que mes por mes se cobra un interés del 0.4% . Llegando a la siguiente ecuación en diferencias:
$$ P_{n} = 1.004(P_{n-1})$$

Con esto, al tomar $P_0 = 694.44$ usando la fórmula de interés compuesto llegamos lo siguiente:
$$  P_{n} = (1.004)^{n}P_0$$

Y ya con esta ecuación conocemos la cantidad que se debe pagar en el mes $n$. Esto funciona sumando los pagos cada mes y luego dividir entre 360, entonces el pago es de $1553.54$ cpara cubrir la deuda a 360 meses



Código de la ecuación en diferencias:

```python
pago_completo = 0 # <---- Variable para ir almacenando la suma de la deuda
p_0 = 250000/360 #<-------   cantidad que se debería pagas mes por mes sin interés
r_i = 1.004 # <----- interés 
for i in range(1,361):
    pago_completo = pago_completo + (math.pow(r_i, i))*p_0
#pago mensual que ya incluye el interés
pago_mensual = pago_completo/360
print(f'Pago mesnual que cubre la deuda en 360 meses : {pago_mensual}')
```

## Modelo SI
Uno de los modelos compartimentales mas simples en epidemiología es el modelo SI (Susceptible-Infectado), presentado por Kermack y McKendrick en 1927. Dado una población de $N$ individuos, suponemos que dada una cantidad inicial $S_0$ de individuos susceptibles a una cierta enfermedad, y una cantidad $I_0$ de individuos infectados con dicha enfermedad, la evolución de la población (en días) durante una epidemia está gobernada por el siguiente sistema de ecuaciones en diferencia
$$
\begin{cases}
\Delta S_n=-\lfloor \beta S_nI_n \rfloor\\
\Delta I_n=\lfloor \beta S_nI_n \rfloor
\end{cases}
$$
 con $n\in \mathbb{N}$, dónde $\beta$ es la tasa de contagio de la enfermedad. 
 - Comente sobre las principales hipotésis que debemos hacer sobre una enfermedad para poder aplicar este modelo. ¿Son razonables? 
 - Utilizando la hipótesis que la población es constante ($S_n + I_n = N$ para todo $n ∈ \mathbb{N}$), escriba este sistema de ecuaciones en diferencia como una única ecuación en diferencia, de la que pueda deducir toda la información. 
 - Comenzando de una población $(S_0,I_0) = (999,1)$ calcule y grafique (de manera bidimensional) el comportamiento de la población en los próximos 100 días si la tasa de contagio es β = 0,1. 
 - El comportamiento a largo plazo, ¿depende de las condiciones iniciales?¿Cuáles son los distintos escenarios que podemos tener luego de mucho tiempo?

###### Comentario sobre las hipótesis del modelo

Lo primero que me viene a la mente de este modelo para que tenga sentido y modele como se comportan las enfermedades infecciosas en una determinada población es que $I_0$ debe ser mayor a cero de lo contrario no habría una enfermedad contagiosa que estudiar. Estoy de acuerdo con lo de población constante ya que este modelo funciona para estudiar cierta población con una cantidad finita de seres vivos donde no pueden surgir más miembros de la población de forma espontánea. Esto nos lleva a que $A$ cantidad de personas están enfermas entonces $N - A$ personas están sanas. Respecto a la tasa de contagio es un supuesto bastante importante por que necesitamos saber que tan rápido o lento la población sana se enferma y viceversa.

######  Conviertiendo el sistema de ecuaciones en diferencias en una única ecuación 


Comenzamos desarrollando las ecuaciones del  sistema de ecuaciones en diferencias por separado.

$$\Delta S_n=-\lfloor \beta S_nI_n \rfloor \implies S_{n+1} - S_{n} = -\lfloor \beta S_nI_n \rfloor $$
$$\Delta I_n=\lfloor \beta S_nI_n \rfloor \implies I_{n+1} - I_{n} = \lfloor \beta S_nI_n \rfloor $$


Como la población es constante entonces $I_{n} +S_{n} = N$ entonces $S_{n} = N - I_{n}$

$$I_{n+1} = \lfloor \beta S_nI_n \rfloor + I_{n} =  \lfloor \beta \left(N - I_{n}\right)I_n \rfloor + I_{n}$$
Con esto llegamos a una única ecuación en diferencias de la cual podemos deducir toda la información 

###### Implementación computacional  de la ecuación en diferencias

```python
N = 1000  # Numero total de individuos
I_0 = 1  # Condición inicial, Ddo que el supuesto de población constante es cierto, no es necesario escribir la variable S_0 = 999
beta = 0.1 # Tsa de contagio 
listaI = [] # En esta lista se irán almacenando los valores que calculamos de acuerdo a la ecuación en diferencias a la que legamos arriba
'''
Con ayuda de un ciclo for realizamos la tarea
El primero if sirve para agregar la condición inicial al arreglo.
Y después solo seguira el flujo del codigo en el else, que es aplicar la operacion de la ecuación en diferencias apartir del elemento anterior 
del arreglo.
Se repite 100 veces por que hace analogia a los 100d días que se estudia el modelo
'''
for i in range(101):
    if (i == 0):
        listaI.append(I_0)
    else:
        listaI.append(math.floor(beta*(N - listaI[i - 1]))  + listaI[i - 1])
listaS = [N-listaI[i] for i in range(len(listaI))]
#Gráficación
fig, ax = plt.subplots(figsize=(6, 3.0), layout='constrained')
plt.scatter(listaI,listaS, s=5.5)
plt.grid()
plt.ylabel('$S_n$')
plt.xlabel('$I_n$')
plt.title('Comportamiento de la Población en los proximos 100 días')
plt.show()
```

**Obervaciones de la gráfica**

A medida que los días pasan $S_ n $ se acerca a 0 e $i_n$ a 1000. Se aprecia que los datos se comportan decrecientemente

Notamos que la condicón inicial de la gráfica es $(999,1)$, y a medida que los días pasas  la $I_k$ aumenta y la $S_k$ disminuye. Notamos que nos llevan a un punto de equilibrio en $(9, 991)$

Veamos que pasa cuando variamos las condiciones inicales. Tomando $(500,500) = (S_0, I_0)$

```python
N = 1000
I_0 = 500
beta = 0.1
listaI = []
for i in range(101):
    if (i == 0):
        listaI.append(I_0)
    else:
        listaI.append(math.floor(beta*(N - listaI[i - 1]))  + listaI[i - 1])
listaS = [N-listaI[i] for i in range(len(listaI))]
fig, ax = plt.subplots(figsize=(6, 3.0), layout='constrained')
plt.scatter(listaI,listaS, s=5.5)
plt.grid()
plt.ylabel('$S_n$')
plt.xlabel('$I_n$')
plt.title('Comportamiento de la Población en los proximos 100 días. Condición inicial diferente')
plt.show()
```

## Modelo de supervivencia

Suponga que estamos estudiando la supervivencia de una cierta especie de ballenas, y sabemos que si el número de ballenas es menor que cierto nivel mínimo de supervivencia $m$, la especie se extingue. Suponga también que la población es limitada por una cierta población máxima $M$. Es decir, si la población de ballenas está por encima de $M$, ésta empezará a descender porque el ecosistema no puede soportar una población tan grande. Podemos modelar esta situación usando la siguente ecuación, donde an representa la población de ballenas luego de $n$ años. $$a_{n+1} −a_n = K(M −a_n)(a_n −m)$$ Discuta el modelo.

```python
m = 10 # Valor mínimo de ballenas, si se llega a ese valor, la especie se extingue 
M = 1000 # Valor máximo de ballenas, si se alcanza este valor, la cantidad emoezará a descender
k = 0.00008 # Tasa de decremento 
a_0 = 80  # condición inicial, cantidad de ballenas con las que se empieza a estudiar al modelo 
```

```python
'''
Este código hace el cáclulo de la ecuación en diferencias para 100 años
es por eso que el for repite la acción 100 veces, cada valor obtenido por año 
se almacena en el arreglo datos
'''
datos = []
for i in range(101):
    if i == 0:
        datos.append(a_0)
    else:
        datos.append(int(datos[i - 1] + k*(M - datos[i - 1])*(datos[i - 1] - m)))
```

```python
#Gráficación 
fig, ax = plt.subplots(figsize=(6, 3.0), layout='constrained')
plt.scatter([i for i in range(101)],datos,s = 2.5, color = 'hotpink')
plt.grid()
plt.ylabel('Número de Ballenas')
plt.xlabel('Años')
plt.title('Modelo de Supervivencia en Población de Ballenas')
plt.show()
```

**Observación**


Notamos que la gráfica descrita arriba es parecido a un modelo logístico acotado en $10$ y $1000$. Describe bastante bien como se comporta la población de ballenas a lo largo de los años, toma en cuenta también la condición inicial establecida y el tope máximo de especies en el ecosistema.
